```python
import numpy as np
import matplotlib.pyplot as plt

# Función de la parábola: f(x) = ax^2 + bx + c
def parabola(x, a, b, c):
    return a * x**2 + b * x + c

# Derivada de la función de la parábola: f'(x) = 2ax + b
def derivada_parabola(x, a, b):
    return 2 * a * x + b

# Coeficientes de la parábola
a = float (input("ingrese a:"))  # coeficiente cuadrático
b = float (input("ingrese b:"))  # coeficiente lineal
c = float (input("ingrese c:"))  # término independiente

# Rango de valores de x
x_values = np.linspace(-10, 10, 1100)

# Evaluar la función de la parábola y su derivada
y_parabola = parabola(x_values, a, b, c)
y_derivada = derivada_parabola(x_values, a, b) #Deriva respecto a los valores de x dados anteriomente, asi que solo hay que declarar los coeficientes de x^2 y x

# Encontrar el máximo de la parábola (vértice)
x_max = -b / (2 * a)
y_max = parabola(x_max, a, b, c)

# Graficar la parábola y su derivada
plt.figure(figsize=(8, 6))

#Ventana que indica la funcion y su derivada
plt.plot(x_values, y_parabola, label='Parábola: f(x) = {}x^2 + {}x + {}'.format(a, b, c))
plt.plot(x_values, y_derivada, label="Derivada: f'(x) = {}x + {}".format(2*a, b))

#declaración del punto, su color y contenido
plt.scatter(x_max, y_max, color='purple', marker='s', label=f'Máximo en ({x_max:.2f}, {y_max:.2f})')

#configuración del plano cartesiano
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

#Nombre de los ejes
plt.xlabel('x')
plt.ylabel('y')

#Declaración del titulo
plt.title('Gráfica de la Parábola y su Derivada')

#Nos genera la ventana  plt.plot
plt.legend()

#Declara que se genere la ventana
plt.grid(True)

#Muestra la ventana en debug
plt.show()
```

```python
import time #medimos cuánto tarda en ejecutarse el programa.
inicio = time.time() #Esta línea guarda el tiempo actual en el cual el programa se ejecuta
print("Evaluación de números primos\n")
a = int(input("Ingrese el numero desde donde iniciara la evaluación:"))
b = int(input("Ingrese el numero desde donde finalizara la evaluación:"))
print("\n")
print("Los números primos son:\n")
for i in range(a,b+1): #El rango de numeros a ser evaluados si son primos
    conta = 0 #Se inicializa la variable que va a mostrar cuantos divisores tiene
    for n in range(1, i+1): # Se  evaluan los divisores de i (excluyendo i)
        residue = i%n
        if residue == 0:
            conta = conta + 1

    if conta == 2: #Un numero primo solo tiene dos divisores el 1 y el mismo
        print(f'{i} es un primo')

fin = time.time() #Guarda el tiempo en el cual finaliza el programa
print("\n")
print("El tiempo de ejecución del programa fue de:", (fin - inicio)*1000)
```
